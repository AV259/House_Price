```python
df1['SALE_COND'].unique()
```

```python
df1['PARK_FACIL'].unique()
```

```python
df1['PARK_FACIL'].replace(['Noo'],'No',inplace=True)
```

```python
df1['BUILDTYPE'].unique()
```

```python
df1['BUILDTYPE'].replace(['Comercial'],'Commercial',inplace=True)
df1['BUILDTYPE'].replace(['Others'],'Other',inplace=True)
```

```python
df1['UTILITY_AVAIL'].unique()
```

```python
df1['UTILITY_AVAIL'].replace(['AllPub'],'All Pub',inplace=True)
```

```python
df1['STREET'].unique()
```

```python
df1['MZZONE'].unique()
```

```python
df1['UTILITY_AVAIL'].replace(['NoSeWa'],'NoSewr',inplace=True)
df1['UTILITY_AVAIL'].replace(['NoSewr '],'NoSewr',inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

Linear Regression 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train,y_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train_scaled,y_train)
```

```python
y_pred = regressor.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score
```

```python
print('R2 Score is : ' , r2_score(y_test,y_pred))
```

Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=200)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
from sklearn import metrics
print('R2 Score is : ' , r2_score(y_test,y_pred))
```

Random Forest is giving us better results with R2 score of 0.989
